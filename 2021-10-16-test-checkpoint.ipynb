# Задание 2

Импортируем нужные библиотеки

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

Апроксимирующая функция

```python
def func(x, a, b):
    return a*np.log(x+1)+b
```

Импортируем данные

```python
df = pd.read_csv(r'C:\Users\anton\OneDrive\Desktop\Проги\Tests\data.csv') 
x = df['day']
y = df['revenue']
y1 = df['plrs']
```

С помощю функции curve_fit рассчитаем коэффициенты для функций суммы платежей и количества игроков.

```python
coeffs1, matcov1 = curve_fit(func, x, y )
yaj1 = func(x, coeffs1[0], coeffs1[1] )
coeffs2, matcov2 = curve_fit(func, x, y1 )
yaj2 = func(x, coeffs2[0], coeffs2[1] )
```

Выведем получившиеся коэффициенты.

```python
print('Сумма платежей', round(coeffs1[0],2),round(coeffs1[1],2))
print('Игроки', round(coeffs2[0],2),round(coeffs2[1],2))
```

Нарисуем график суммы платежей.

```python
plt.xlabel('days')
plt.ylabel('revenue')
plt.plot( x, y, 'x', x, yaj1 ,'r-' )
plt.show()
```

Нарисуем график кол-ва игроков.

```python
plt.plot( x, y1, 'x', x, yaj2 ,'r-' )
plt.xlabel('days')
plt.ylabel('paying players')
plt.show()
```

Рассчитаем ARPPU 15, 30 и 60 дня.

```python
print('day15 ARPPU', round(func(15,coeffs1[0],coeffs1[1])/func(15,coeffs2[0],coeffs2[1]),2))
print('day30 ARPPU', round(func(30,coeffs1[0],coeffs1[1])/func(30,coeffs2[0],coeffs2[1]),2))
print('day60 ARPPU', round(func(60,coeffs1[0],coeffs1[1])/func(60,coeffs2[0],coeffs2[1]),2),'\n')
```

А теперь рассчитаем ARPU 15, 30 и 60 дня.

```python
print('day15 ARPU', round(func(15,coeffs1[0],coeffs1[1])/69000,2))
print('day30 ARPU', round(func(30,coeffs1[0],coeffs1[1])/69000,2))
print('day60 ARPU', round(func(60,coeffs1[0],coeffs1[1])/69000,2),'\n')
```
